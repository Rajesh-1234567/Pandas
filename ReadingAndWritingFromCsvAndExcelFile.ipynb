```python
# Read Excel

df = pd.read_excel("stock_data.xlsx","Sheet1")
df
```

```python
def convert_people_cell(cell): # created a function which converts prople column where na presents and accepts a parameter as cell
    if cell=="n.a.":
        return 'Sam Walton'
    return cell

def convert_price_cell(cell):# also same for price column
    if cell=="n.a.":
        return 50
    return cell
    
df = pd.read_excel("stock_data.xlsx","Sheet1", converters= { # using converters we can convert
        'people': convert_people_cell,
        'price': convert_price_cell
    })
df
```

```python
#. Write to Excel
df.to_excel("new.xlsx", sheet_name="stocks", index=False, startrow=2, startcol=1) # addind the starting row and col
df
```

```python
# Write two dataframes to two separate sheets in excel

df_stocks = pd.DataFrame({
    'tickers': ['GOOGL', 'WMT', 'MSFT'],
    'price': [845, 65, 64 ],
    'pe': [30.37, 14.26, 30.97],
    'eps': [27.82, 4.61, 2.12]
})

df_weather =  pd.DataFrame({
    'day': ['1/1/2017','1/2/2017','1/3/2017'],
    'temperature': [32,35,28],
    'event': ['Rain', 'Sunny', 'Snow']
})
```
